```python
init_date = pd.to_datetime('20160114')
iter_times = 50
iter_interval = 10
top_k = 5
```

```python
def strategy(income_df: pd.DataFrame, period : int, cash: float, iter_times: int, iter_interval : int, top_k : int) -> (list,dict):
    stock_in_hand = {}
    stock_in_hand_all_history = {}
    
    iter_date = init_date
    assets = []
    for i in range(iter_times):
        iter_df = income_df[income.df.date == iter_date]
        top_k_df = iter_df.sort_values(by=['income'], ascending=False)[:top_k]
        
        # if no data at that time, the trade will be delivered on the next day
        if top_k_df.empty:
            iter_date +=pd.DateOffset(1)
        continue
        
        # sell stocks not in top_k list
        for stock in stock_in_hand.copy():
            if not stock in top_k_df['ticker'].tolist():
                cash += iter_df[iter_df.ticker == stock]['close'].value[0] * stock_in_hand[stock] 
                stock_in_hand.pop(stock)
        
        # compute how many new stocks to buy
        new_to_buy = 0 
        for stock in top_k_df.ticker.tolist():
            if not stock in stock_in_hand:
                new_to_buy +=1
        cash_on_each = cash / new_to_buy
        
        # buy new top_k stocks 
        for stock in top_k_df.ticker.tolist():
            if not stock in stock_in_hand:
                stock_in_hand[stock] = cash_on_each / iter_df[iter_df.ticker == stock]
                cash -= cash_on_each
                
        # compute current assets
        cur_asstes = cash
        for stock in stock_in_hand:
            cur_assets += iter_df[iter_df.ticker == stock]['close'].values[0] * stock_in_hand[stock]
            
        assets.append(cur_assets)
        stock_in_hand_all_history[iter_date] = stock_in_hand.copy()
        iter_date +=pd.DateOffset(iter_interval)
        
    return assets, stock_in_hand_all_history
            
```
