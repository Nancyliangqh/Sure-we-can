```python
import datetime
import numpy as np
import pandas as pd
%matplotlib inline
pd.core.common.is_list_like = pd.api.types.is_list_like
import pandas_datareader.data as web
import fix_yahoo_finance as yf
```

```python
def get_data(stock,start_date,end_date):
    yf.pdr_override()
    dataframe = web.get_data_yahoo(stock, start = start_date, end = end_date)
    dataframe = dataframe[['Close', 'Volume']]
    # Rename column names
    close = dataframe['Close']
    volume = dataframe['Volume']
    clolist = close.columns.tolist()
    adjclolist = [x + "_Close" for x in clolist]
    close.columns = adjclolist
    vollist = volume.columns.tolist()
    adjvollist = [x + "_Volume" for x in vollist]
    volume.columns = adjvollist
    df = pd.concat([close, volume], axis = 1, sort = False)
    df = df.dropna(axis = 'columns')
    return df
```
