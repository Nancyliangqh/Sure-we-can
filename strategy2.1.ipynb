```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('SampleData.csv')
columns = df.columns.tolist()

for x in columns[1:26]:
    index = columns.index(x)
    if index < 26:
        market = df[df.columns[index]] * df[df.columns[index+25]]
        df[str(x) + '_' + 'Mkt'] = market
df = df.set_index('Date')
df_small = df.iloc[:, 50:76]# choose data
```

```python
# create new data frame named returns for strategy2
small = df_small.T #transfer DataFrame
columns = small.columns.tolist()
for i in columns[0:504]:
    index = columns.index(i)
    small[str(i) + '_' + 'returns'] = (small[small.columns[index+1]] - small[small.columns[index]]) / small[small.columns[index]]
df_returns = small.iloc[:, 504:] # use market values to calculate returns every day
```

```python
# create portfolio_list2 for calculating returns for strategy2
# choose top 5 stocks whose returns are largest to long
def func_portfolio_list2():
    buy2 = []
    for i in range (len(df_returns.columns)):
        b = df_returns
        large_returns = b.sort_values([b.columns[i]],ascending=False) # sort each column and choose top 5 stocks every day
        ones = np.ones(5, dtype=int)
        choice = np.array(large_returns.index[0:5]).tolist()
        c = dict(zip(choice, ones))# create a new dictionary to show stocks longed
        buy2.append(c)
        
    # 'portfolio2' is same as 'buy2' on the first day(buy top 5 stocks whose returns are largest)
        portfolio_list2 = [buy2[0]]
        for i in range(len(buy2) - 1):
            dictMerged = dict(buy2[i], **buy2[i+1]) # merge dictionarys
            for j in buy2[i]:
                if j in buy2[i+1]: # if stock A exists in the both portfolios, we set it to be 0 in the merged portfolio
                    dictMerged[j] = 0 # 0 means no actions
                
                else:            # if stock A exists in the first portfolio but not in the second portfolio, we set it to be -1 in the merged portfolio
                    dictMerged[j] = -1 # -1 means shorting the stock
                
                             # if stock A exists in the second portfolio but not in the first portfolio, we set it to be 1 in the merged portfolio
                             # 1 means longing the stock(same as 'buy')
                             
            portfolio_list2.append(dictMerged)
    return portfolio_list2 # getting portfolio_list2 during the trading days
func_portfolio_list2()
```
