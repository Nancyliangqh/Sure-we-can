```python
# create portfolio_list2 for calculating returns for strategy2
# choose top 5 stocks whose returns are largest to long
def func_portfolio_list2():
    buy2 = []
    for i in range (len(df_returns.columns)):
        b = df_returns
        large_returns = b.sort_values([b.columns[i]],ascending=False) # sort each column and choose top 5 stocks every day
        ones = np.ones(5, dtype=int)
        choice = np.array(large_returns.index[0:5]).tolist()
        c = dict(zip(choice, ones))# create a new dictionary to show stocks longed
        buy2.append(c)
        
    # 'portfolio2' is same as 'buy2' on the first day(buy top 5 stocks whose returns are largest)
        portfolio_list2 = [buy2[0]]
        for i in range(len(buy2) - 1):
            dictMerged = dict(buy2[i], **buy2[i+1]) # merge dictionarys
            for j in buy2[i]:
                if j in buy2[i+1]: # if stock A exists in the both portfolios, we set it to be 0 in the merged portfolio
                    dictMerged[j] = 0 # 0 means no actions
                
                else:            # if stock A exists in the first portfolio but not in the second portfolio, we set it to be -1 in the merged portfolio
                    dictMerged[j] = -1 # -1 means shorting the stock
                
                             # if stock A exists in the second portfolio but not in the first portfolio, we set it to be 1 in the merged portfolio
                             # 1 means longing the stock(same as 'buy')
                             
            portfolio_list2.append(dictMerged)
    return portfolio_list2 # getting portfolio_list2 during the trading days
```

```python
# return a dictionary that shows how many shares for each stock should be bought on the first day
def first_day(x, cashpool, stocknum):
    global df_returns
    position = df_returns.columns.tolist().index(x) # get the column index of the date x input
    df_returns = df_returns.sort_values([x],ascending=False) # sort according to date x
    choice = np.array(df_returns.index[0:stocknum]).tolist() # get the largest N(stocknum) stocks of the day
    choice = [i.split('_')[0] for i in choice]

    # get the column name I want to find the close price for each stock on day x
    price = [i.split('_')[0] + "_" + 'Close' for i in choice]

    # find the close price for each stock on day x
    close_price = []
    for i in range(len(price)):
        b = df.loc[x, price[i]]
        close_price.append(b) # get the close price for stocks chosen from df
        invest = cashpool / stocknum
        share = [int(invest / i) for i in close_price] # get the shares for stocks chosen
        buy_dict = dict(zip(choice, share)) # update buy_list for INITIAL day x
    return buy_dict
first_day('2016-11-21',10000,5)
```

```python
# create function allocation to return 3 dictionaries showing the change(buy sell keep) for stocks on the second day
def allocation(df, df_returns, x, y, cashpool, stocknum ): # date x -- enter stock market; y- date want to make change
    position = df_returns.columns.tolist().index(x) # get the column index of the date x input
    df_returns = df_returns.sort_values([x],ascending=False) # sort according to date x(descending)
    choice = np.array(df_returns.index[0:stocknum]).tolist() # get the largest N(stocknum) stocks of the day x
    choice = [i.split('_')[0] for i in choice]

    # get the column name I want to find the close price for each stock on day x
    price = [i.split('_')[0] + "_" + 'Close' for i in choice]

    # find the close price for each stock on day x
    close_price = []
    for i in range(len(price)):
        b = df.loc[x, price[i]]
        close_price.append(b) # get the close price for stocks chosen from df
        invest = cashpool / stocknum
        share = [int(invest / i) for i in close_price] # get the shares for stocks chosen
        buy_dict = dict(zip(choice, share)) # update buy_list for INITIAL day x
    
    def new_basket(df_returns, y): # given day y - that we want to get portfolio change
        position = df_returns.columns.tolist().index(y) # get the column index of second day of the date y input
        df_returns2 = df_returns.sort_values([y], ascending=False) # sort according to date y(descending)
        choice2 = np.array(df_returns2.index[0:5]).tolist() # get the largest N(stocknum) stocks of the day y 
        choice2 = [i.split('_')[0] for i in choice2]
        ones = np.ones(5, dtype=int) # give all 1 to buy for each stock
        a = dict(zip(choice2, ones)) # buy basket for date y
        return a

    ### FINALIZED 3 dicts to return at the end ###
    a = new_basket(df_returns,y)
    buy = {k:a[k] for k in a if k not in buy_dict} # in y day's purchase list, not in day x, buy(exact shares will be calculated by the class(securities))
    keep = {k:0 for k in a if k in buy_dict} # in both days' purchase list, no change
    sell = {k:-buy_dict[k] for k,v in buy_dict.items() if k not in a} # in day x not in y day's purchase, sell all shares

    return (buy, keep, sell)
```
